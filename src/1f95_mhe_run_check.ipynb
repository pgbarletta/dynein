```julia
using DelimitedFiles
using Plots
```

# 1F95

## minimization

```julia
dir_f95_min="/home/pbarletta/labo/20/dynein/run/1f95/mhe/m"
f95_min_Etot = readdlm("$dir_f95_min/outputs/Etot_min")[:, 2]
f95_mmin_Etot = readdlm("$dir_f95_min/outputs/Etot_mmin")[:, 2];

ntpr = 5
NSTEPS = ntpr * (length(f95_min_Etot) + length(f95_min_Etot))
x = collect(0:5:NSTEPS-1)

plot(x, [ f95_min_Etot ; f95_mmin_Etot],
    label = "1F95", title = "Solute + Solvent minimization",
    yaxis = "E", xaxis = "NSTEP")
```

## heat

```julia
# Energies
dir_f95_hea="/home/pbarletta/labo/20/dynein/run/1f95/mhe/h"
f95_hea_Ek = readdlm("$dir_f95_hea/outputs/ektot.dat")[:, 2]
f95_hea_Ep = readdlm("$dir_f95_hea/outputs/eptot.dat")[:, 2]
f95_hea_Etot = readdlm("$dir_f95_hea/outputs/etot.dat")[:, 2]

plot(f95_hea_Ek,
    label = "1F95_K", title = "Energa - Calentamiento", size = (750, 400),
    yaxis = "E", xaxis = "NSTEP")
plot!(f95_hea_Ep, label = "1F95_U")
plot!(f95_hea_Etot, label = "1F95_E")
```

```julia
f95_hea_T = readdlm("$dir_f95_hea/outputs/temp.dat")[:, 2];

plot(f95_hea_T,
    label = "1F95", title = "Temperatura - Calentamiento", size = (750, 400),
    yaxis = "Temperature", xaxis = "NSTEP")
```

## equilibration

```julia
# Energies
dir_f95_equ="/home/pbarletta/labo/20/dynein/run/1f95/mhe/e"
f95_equ_Ek = readdlm("$dir_f95_equ/outputs/ektot.dat")[:, 2]
f95_equ_Ep = readdlm("$dir_f95_equ/outputs/eptot.dat")[:, 2]
f95_equ_Etot = readdlm("$dir_f95_equ/outputs/etot.dat")[:, 2]

plot(f95_equ_Ek,
    label = "1F95", title = "Energa - Equilibracin", size = (750, 400),
    xlims = (15E3, 20E3), #ylims = (-3E5, -2.45E5),
    yaxis = "E", xaxis = "NSTEP")
plot!(f95_equ_Ep, label = "1F95_U")
plot!(f95_equ_Etot, label = "1F95_E")
```

```julia
# Temperature
f95_equ_T = readdlm("$dir_f95_equ/outputs/temp.dat")[:, 2]

plot(f95_equ_T,
    label = "1F95", title = "Temperatura - Equilibracin", size = (750, 400),
    yaxis = "Temperature", xaxis = "NSTEP")
```

```julia
# Density
lb3_equ_dens = readdlm("$dir_f95_equ/outputs/density.dat")[:, 2]

plot(lb3_equ_dens,
    label = "1F95", title = "Densidad - Equilibracin", size = (750, 400),
    yaxis = "Density", xaxis = "NSTEP")
```

```julia
# RMSD ntwx=500
f95_equ_rmsd = readdlm("$dir_f95_equ/rmsd_equ_1f95")[2:end, 2]

plot(f95_equ_rmsd,
    xlims = (1, 2E4), xticks = 0:2E3:20E3, yticks = 0:.2:1.5,
    label = "1F95", title = "RMSD - Equilibracin", size = (750, 400),
    yaxis = "RMSD", xaxis = "NSTEP")
```
