# PCA projección
#### Obtengo vectores diferencia de la trayectoria. Esto lo hago una sola vez

#### Obtengo las proyecciónes de vectores diferencia sobre los 1eros PCA

```julia
dif = h5read(joinpath(rtdos, string("diff_", pdb, ".h5")), "dif")
norm_dif = mapslices(x -> x ./ norm(x), dif, dims = 1)
nframes = size(dif)[2]

prj_1_f96 = Array{Float64, 1}(undef, nframes)
prj_2_f96 = Array{Float64, 1}(undef, nframes)
prj_3_f96 = Array{Float64, 1}(undef, nframes)

for i = 1:nframes
    prj_1_f96[i] = dot(norm_dif[:, i], modes_f96[:, 1])
    prj_2_f96[i] = dot(norm_dif[:, i], modes_f96[:, 2])
    prj_3_f96[i] = dot(norm_dif[:, i], modes_f96[:, 3])
end

# Smooth
smooth_prj_1_f96 = [ mean(prj_1_f96[i:i+9]) for i = 1:10:length(prj_1_f96)-9 ]
smooth_prj_2_f96 = [ mean(prj_2_f96[i:i+9]) for i = 1:10:length(prj_2_f96)-9 ]
smooth_prj_3_f96 = [ mean(prj_3_f96[i:i+9]) for i = 1:10:length(prj_3_f96)-9 ];
```

```julia
plt_2his = histogram2d(smooth_prj_1_f96, smooth_prj_2_f96, 
    xlims = (-1, 1), ylims = (-1, 1),
    xticks = [-.9 ; -.5 ; 0. ;.5 ; .9], yticks = [-.9 ; -.5 ; 0. ;.5 ; .9],
    yaxis = "PCA 2", xaxis = "PCA 1",
    colorbar_title = "Frame count",
    guidefont = font(22, "Arial"), tickfont = font(16, "Arial"),
    legendfont = font(16, "Arial"), #clims = (0, 15),
    nbins = 120, seriescolor = :blues, grid = false)
```

```julia
plt_2his = histogram2d(smooth_prj_2_f96, smooth_prj_3_f96, 
    xlims = (-1, 1), ylims = (-1, 1),
    xticks = [-.9 ; -.5 ; 0. ;.5 ; .9], yticks = [-.9 ; -.5 ; 0. ;.5 ; .9],
    yaxis = "PCA 3", xaxis = "PCA 2",
    colorbar_title = "Frame count",
    guidefont = font(22, "Arial"), tickfont = font(16, "Arial"),
    legendfont = font(16, "Arial"), clims = (0, 15),
    nbins = 120, seriescolor = :blues, grid = false)
```
