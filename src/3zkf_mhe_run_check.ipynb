# 3ZKF

## minimization

```julia
dir_zkf_min="/home/pbarletta/labo/20/dynein/run/3zkf/mhe/m"
zkf_min_Etot = readdlm("$dir_zkf_min/outputs/Etot_min")[:, 2]
zkf_mmin_Etot = readdlm("$dir_zkf_min/outputs/Etot_mmin")[:, 2];

ntpr = 5
NSTEPS = ntpr * (length(zkf_min_Etot) + length(zkf_min_Etot))
x = collect(0:5:NSTEPS-1)

plot(x, [ zkf_min_Etot ; zkf_mmin_Etot],
    label = "3ZKF", title = "Solute + Solvent minimization",
    yaxis = "E", xaxis = "NSTEP")
```

## heat

```julia
# Energies
dir_zkf_hea="/home/pbarletta/labo/20/dynein/run/3zkf/mhe/h"
zkf_hea_Ek = readdlm("$dir_zkf_hea/outputs/ektot.dat")[:, 2]
zkf_hea_Ep = readdlm("$dir_zkf_hea/outputs/eptot.dat")[:, 2]
zkf_hea_Etot = readdlm("$dir_zkf_hea/outputs/etot.dat")[:, 2]

plot(zkf_hea_Ek,
    label = "3ZKF_K", title = "Energa - Calentamiento", size = (750, 400),
    yaxis = "E", xaxis = "NSTEP")
plot!(zkf_hea_Ep, label = "3ZKF_U")
plot!(zkf_hea_Etot, label = "3ZKF_E")
```

```julia
zkf_hea_T = readdlm("$dir_zkf_hea/outputs/temp.dat")[:, 2];

plot(zkf_hea_T,
    label = "3ZKF", title = "Temperatura - Calentamiento", size = (750, 400),
    yaxis = "Temperature", xaxis = "NSTEP")
```

## equilibration

```julia
# Energies
dir_zkf_equ="/home/pbarletta/labo/20/dynein/run/3zkf/mhe/e"
zkf_equ_Ek = readdlm("$dir_zkf_equ/outputs/ektot.dat")[:, 2]
zkf_equ_Ep = readdlm("$dir_zkf_equ/outputs/eptot.dat")[:, 2]
zkf_equ_Etot = readdlm("$dir_zkf_equ/outputs/etot.dat")[:, 2]

plot(zkf_equ_Ek,
    label = "3ZKF", title = "Energa - Equilibracin", size = (750, 400),
    xlims = (15E3, 20E3), #ylims = (-3E5, -2.45E5),
    yaxis = "E", xaxis = "NSTEP")
plot!(zkf_equ_Ep, label = "3ZKF_U")
plot!(zkf_equ_Etot, label = "3ZKF_E")
```

```julia
# Temperature
zkf_equ_T = readdlm("$dir_zkf_equ/outputs/temp.dat")[:, 2]

plot(zkf_equ_T,
    label = "3ZKF", title = "Temperatura - Equilibracin", size = (750, 400),
    yaxis = "Temperature", xaxis = "NSTEP")
```

```julia
# Density
lb3_equ_dens = readdlm("$dir_zkf_equ/outputs/density.dat")[:, 2]

plot(lb3_equ_dens,
    label = "3ZKF", title = "Densidad - Equilibracin", size = (750, 400),
    yaxis = "Density", xaxis = "NSTEP")
```

```julia
# RMSD ntwx=500
zkf_equ_rmsd = readdlm("$dir_zkf_equ/rmsd_equ_3zkf")[2:end, 2]

plot(zkf_equ_rmsd,
    xlims = (1, 2E4), xticks = 0:2E3:20E3, yticks = 0:.2:1.5,
    label = "3ZKF", title = "RMSD - Equilibracin", size = (750, 400),
    yaxis = "RMSD", xaxis = "NSTEP")
```
